## 실습 01 네이버주가크롤링
### 1. 품목명과현재가를크롤링해주세요.

```python
URL = 'https://finance.naver.com/sise/sise_quant.nhn'
request_headers = { 
'User-Agent' : ('Mozilla/5.0 (Windows NT 10.0;Win64; x64)\
AppleWebKit/537.36 (KHTML, like Gecko) Chrome/71.0.3578.98\
Safari/537.36'), }
response = requests.get(URL,headers = request_headers)
soup = BeautifulSoup(response.text, "html.parser")
```

```python
result = soup.find("table", class_= "type_2")
result_tr = result.find_all('tr')

# result_tr = soup.select("table.type_2 tr")
```

```python
for i in result_tr:
    if i.find('a') is not None:
        print(i.find('a').text, i.find('td',class_="number").text)
```

## 실습 02 무신사 쇼핑몰의 상의(TOP) 카테고리 첫페이지의 제품들의 브랜드명, 제품명, 가격을 아래와 같이 크롤링해주세요.


```python
URL = 'https://search.musinsa.com/category/001'
request_headers = { 
'User-Agent' : ('Mozilla/5.0 (Windows NT 10.0;Win64; x64)\
AppleWebKit/537.36 (KHTML, like Gecko) Chrome/71.0.3578.98\
Safari/537.36'), }
response = requests.get(URL,headers = request_headers)
soup = BeautifulSoup(response.text, "html.parser")
result_li = soup.select("li.li_box")

for i in result_li:
    if i.find('p',class_='price') is not None:        
        temp = re.sub('\n','',i.find_all('p')[1].text)
        if i.find_all('p')[2].text.split(' ')[1] == "":
            print('브렌드 : ', i.find_all('p')[0].text, '\n제품명:', re.sub(' +',' ',temp).lstrip(), '\n가격 : ',i.find_all('p')[2].text.strip(),"\n")
        else:
            print('브렌드 : ', i.find_all('p')[0].text, '\n제품명:', re.sub(' +',' ',temp).lstrip(), '\n가격 : ',i.find_all('p')[2].text.split(' ')[1].strip(),"\n") 
```

```python
#강사님이 하신거

URL = 'https://search.musinsa.com/category/001'
request_headers = { 
'User-Agent' : ('Mozilla/5.0 (Windows NT 10.0;Win64; x64)\
AppleWebKit/537.36 (KHTML, like Gecko) Chrome/71.0.3578.98\
Safari/537.36'), }
response = requests.get(URL,headers = request_headers)
soup = BeautifulSoup(response.text, "html.parser")
article_infos = soup.select("#searchList div.article_info")

for article_info in article_infos:
    print("브렌드 :",article_info.select_one('p.item_title').text.strip())
    print("제품명 :",article_info.select_one('p.list_info > a').attrs['title'])
    print("가격 :", article_info.select_one('p.price').text.strip().split()[-1])
```
